### To remove Outliers.

```python
# Calculating Z_Scores of the Cleaned DataFrame.
z_scores=stats.zscore(cred_card)
```

```python
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
```

```python
cred_card_final=cred_card[filtered_entries]
cred_card_final
```

```python
print(cred_card_final.shape)
print(cred_card.shape)
```

```python
# Defining Dependent & Independent Variables
X=cred_card_final.iloc[:,:-1].values
y=cred_card_final.iloc[:,-1].values
```

```python
# Importing Required Packages for Modeling
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,f1_score,ConfusionMatrixDisplay,confusion_matrix,precision_recall_curve,roc_auc_score,plot_roc_curve,plot_precision_recall_curve,RocCurveDisplay,roc_curve
```

```python
# Training the DataFrame (random_state=3)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=3)
```

```python
# Shape of X_train DataFrame
print(X_train.shape)

# Shape of y_train DataFrame
print(y_train.shape)
```

```python
# Importing Logistic Regression Algorithm
from sklearn.linear_model import LogisticRegression
```

```python
# Defining the Algorithm in Variable form
model=LogisticRegression()
# Fitting the model
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
# Confusion Matrix
pd.crosstab(y_test,y_predict)
```

```python
# Accuracy of training Dataset
model.score(X_train,y_train)
```

```python
# Accuracy of testing Dataset
model.score(X_test,y_test)
```

```python
sb.distplot(y_test-y_predict)
```

```python
# Defining the Algorithm with Hyper Parameters values and fitting the model
model=RandomForestClassifier(max_features='auto',max_depth=200,n_estimators=800,min_samples_split=2,random_state=5)
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
# Accuracy of the model
accuracy_score(y_test,y_predict)
```

```python
# Confusion Matrix
pd.crosstab(y_test,y_predict)
```

```python
# Accuracy of training & testing Dataset
print("Accuracy on training set: %f" %model.score(X_train,y_train))
print("Accuracy on test set: %f" %model.score(X_test,y_test))
```

```python
# classification_report
print(classification_report(y_test,y_predict))
```

```python
sb.distplot(y_test-y_predict)
```

## K-Fold Cross_Validation.

```python
# 10 k-fold Cross_Validation. 
from sklearn.model_selection import cross_val_score # instead of train_test_split
scores = cross_val_score(model,X,y,cv=10,scoring='accuracy') # cv means k-fold
print(scores)
print("Average Accuracy:",scores.mean()) # Average Accuracy of the model.

print("Std :",scores.std()) # Standard Deviation of the model after k_fold CV, Range should be in between 0.01-0.02
```
